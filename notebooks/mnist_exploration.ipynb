```python
import numpy as np
import matplotlib.pyplot as plt
import wandb

from src.utils.data_loader import load_dataset

# Load MNIST (train/val/test; weâ€™ll use train+val for examples)
X_train, y_train, X_val, y_val, X_test, y_test_onehot, y_test_labels = load_dataset("mnist")

X_samples = np.vstack([X_train, X_val])
y_samples = np.vstack([y_train, y_val])
labels = np.argmax(y_samples, axis=1)
```

```python
num_classes = 10
num_per_class = 5

indices_per_class = []
for c in range(num_classes):
    idx = np.where(labels == c)[0]
    indices_per_class.append(idx[:num_per_class])
```

```python
wandb.login()  # only first time
run = wandb.init(project="DA6401_Assignment1", name="mnist_data_exploration")
```

```python
table = wandb.Table(columns=["class", "image"])

for c in range(num_classes):
    for idx in indices_per_class[c]:
        img = X_samples[idx].reshape(28, 28)
        table.add_data(int(c), wandb.Image(img))

wandb.log({"mnist_samples_per_class": table})
run.finish()
```
