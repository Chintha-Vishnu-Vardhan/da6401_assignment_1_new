```python
import numpy as np
import matplotlib.pyplot as plt
import wandb
from src.utils.data_loader import load_dataset
```

```python
"""
Section 2.1: Data Exploration and Class Distribution
Logs a 5x10 table of sample images to Weights & Biases.
"""


def main():
    print("Initializing W&B for Data Exploration...")
    # Initialize a specific run for this task
    wandb.init(project="DA6401__Intro_to_DL_Assignment1", name="2.1_Data_Exploration")

    print("Loading MNIST dataset...")
    data = load_dataset("mnist")
    X_train, y_train = data[0], data[1]
    
    # Convert one-hot to labels
    y_labels = np.argmax(y_train, axis=1)

    # Setup the W&B Table columns (0 through 9)
    columns = [f"Digit {i}" for i in range(10)]
    table = wandb.Table(columns=columns)

    # Collect exactly 5 images for each of the 10 classes
    class_images = {i: [] for i in range(10)}
    for img, label in zip(X_train, y_labels):
        if len(class_images[label]) < 5:
            # Reshape the flat 784 array back into a 28x28 image for W&B
            class_images[label].append(wandb.Image(img.reshape(28, 28)))
            
        # Stop searching if we have 5 images for every class
        if all(len(imgs) == 5 for imgs in class_images.values()):
            break

    # Add the images to the table row by row
    print("Populating W&B Table...")
    for row_idx in range(5):
        row = [class_images[cls][row_idx] for cls in range(10)]
        table.add_data(*row)

    # Log the table to your dashboard
    wandb.log({"MNIST_Sample_Images": table})
    wandb.finish()
    print("Successfully logged image table to W&B!")

if __name__ == "__main__":
    main()
```
